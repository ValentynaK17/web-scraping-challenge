```python
# Import relevant libraries
from splinter import Browser
from bs4 import BeautifulSoup as soup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# converts defined columns(based on order of columns)of a dataframe into specified data type (o_type)
def convert_to(df,start,end,o_type):
    df_columns=df.columns
    if start<end:
        for i in range(start,end):
            df[df_columns[i]]=df[df_columns[i]].astype(o_type)
    elif start==end:
        df[df_columns[start]]=df[df_columns[start]].astype(o_type)
    else: print("wrong input of columns order in convert_to function")
```

```python
browser = Browser('chrome')
```

### Step 1: Visit the Website

Use automated browsing to visit the [Mars Temperature Data Site](https://static.bc-edx.com/data/web/mars_facts/temperature.html). Inspect the page to identify which elements to scrape.

   > **Hint** To identify which elements to scrape, you might want to inspect the page by using Chrome DevTools to discover whether the table contains usable classes.


```python
# Visit the website
# https://static.bc-edx.com/data/web/mars_facts/temperature.html
url = "https://static.bc-edx.com/data/web/mars_facts/temperature.html"
browser.visit(url)
```

### Step 2: Scrape the Table

Create a Beautiful Soup object and use it to scrape the data in the HTML table.

Note that this can also be achieved by using the Pandas `read_html` function. However, use Beautiful Soup here to continue sharpening your web scraping skills.

```python
# Create a Beautiful Soup Object
html = browser.html
html_objct = soup(html, 'html.parser')
```

```python
# Extract all rows of data
table_rows=html_objct.select("table.table tr")

# Preview few extracted rows
table_rows[0:3]
```

```python
# parse the header (table within the page has one-row header)
headers=table_rows[0].select("th") 
headers
```

```python
data_rows=table_rows[1:]
# Preview few rows
data_rows[0:2]
```

### Step 3: Store the Data

Assemble the scraped data into a Pandas DataFrame. The columns should have the same headings as the table on the website. Hereâ€™s an explanation of the column headings:

* `id`: the identification number of a single transmission from the Curiosity rover
* `terrestrial_date`: the date on Earth
* `sol`: the number of elapsed sols (Martian days) since Curiosity landed on Mars
* `ls`: the solar longitude
* `month`: the Martian month
* `min_temp`: the minimum temperature, in Celsius, of a single Martian day (sol)
* `pressure`: The atmospheric pressure at Curiosity's location

```python
# Create an empty list
weather_per_date_list=[]

# Loop through the scraped data to create a list of rows
# assumption: data is cleaned and table has no empty cells
headers_text=[i.text.strip() for i in headers]

for i in data_rows:
    data=i.select("td")
    temp_dict={}
    for j, value_html in enumerate(data):
        header=headers_text[j]
        value=value_html.text.strip()
        temp_dict[header]=value
    weather_per_date_list.append(temp_dict)

# Preview few elements
weather_per_date_list[0:2]
```

```python
# Create a Pandas DataFrame by using the list of rows and a list of the column names
mars_temperature_data_df=pd.DataFrame(weather_per_date_list)
```

```python
# Confirm DataFrame was created successfully
mars_temperature_data_df.head(5)
```

### Step 4: Prepare Data for Analysis

Examine the data types that are currently associated with each column. If necessary, cast (or convert) the data to the appropriate `datetime`, `int`, or `float` data types.

  > **Hint** You can use the Pandas `astype` and `to_datetime` methods to accomplish this task.


```python
# Examine data type of each column
mars_temperature_data_df.dtypes
```

```python
# Change data types for data analysis
convert_to(mars_temperature_data_df,2,5,"int64")
convert_to(mars_temperature_data_df,5,7,"float64")
mars_temperature_data_df["terrestrial_date"]=pd.to_datetime(mars_temperature_data_df["terrestrial_date"],format="%Y-%m-%d")
```

```python
# Confirm type changes were successful by examining data types again
mars_temperature_data_df.dtypes
```

### Step 5: Analyze the Data

Analyze your dataset by using Pandas functions to answer the following questions:

1. How many months exist on Mars?
2. How many Martian (and not Earth) days worth of data exist in the scraped dataset?
3. What are the coldest and the warmest months on Mars (at the location of Curiosity)? To answer this question:
    * Find the average the minimum daily temperature for all of the months.
    * Plot the results as a bar chart.
4. Which months have the lowest and the highest atmospheric pressure on Mars? To answer this question:
    * Find the average the daily atmospheric pressure of all the months.
    * Plot the results as a bar chart.
5. About how many terrestrial (Earth) days exist in a Martian year? To answer this question:
    * Consider how many days elapse on Earth in the time that Mars circles the Sun once.
    * Visually estimate the result by plotting the daily minimum temperature.


```python
# 1. How many months are there on Mars?
mars_temperature_data_df["month"].value_counts().sort_index()
```

```python
# 2. How many Martian days' worth of data are there?
count_of_Martian_days=mars_temperature_data_df["sol"].nunique()
count_of_Martian_days
```

```python
# 3. What is the average low temperature by month?
avg_min_temperature_by_month=mars_temperature_data_df.groupby("month")["min_temp"].mean().sort_index()
avg_min_temperature_by_month
```

```python
# Plot the average temperature by month
avg_min_temperature_by_month.plot(kind="bar")
plt.title("Average Min Daily Temperature per Martian Month")
plt.ylabel("Temperature in Celsius")
plt.xlabel("Martian Month")
plt.xticks(rotation=0)
plt.show()
```

```python
# Identify the coldest and hottest months in Curiosity's location
avg_min_temperature_by_month=avg_min_temperature_by_month.sort_values()
coldest_month=avg_min_temperature_by_month.index[0]
coldest_month
```

```python
hottest_month=avg_min_temperature_by_month.index[-1]
hottest_month
```

```python
# Plot the average temperature by month, sorted by values
avg_min_temperature_by_month.plot(kind="bar")
plt.title("Average Min Daily Temperature per Martian Month")
plt.ylabel("Temperature in Celsius")
plt.xlabel("Martian Month")
plt.xticks(rotation=0)
plt.show()
```

```python
# 4. Average pressure by Martian month
avg_pressure_by_month=mars_temperature_data_df.groupby("month")["pressure"].mean().sort_values()
avg_pressure_by_month
```

```python
# Plot the average pressure by month
avg_pressure_by_month.plot(kind="bar")
plt.title("Average Daily Atmospheric Pressure per Martian Month")
plt.ylabel("Atmospheric pressure")
plt.xlabel("Martian Month")
plt.xticks(rotation=0)
plt.show()
```

```python
month_with_lowest_atmospheric_pressure=avg_pressure_by_month.index[0]
month_with_lowest_atmospheric_pressure
```

```python
month_with_highest_atmospheric_pressure=avg_pressure_by_month.index[-1]
month_with_highest_atmospheric_pressure
```

```python
# 5. How many terrestrial (earth) days are there in a Martian year?
# Add count of Earth Days
mars_temperature_data_df=mars_temperature_data_df.sort_values(by="terrestrial_date")
mars_temperature_data_df.head(5)
```

```python
start_date=mars_temperature_data_df.loc[0,"terrestrial_date"]
mars_temperature_data_df_copy=mars_temperature_data_df.copy()
mars_temperature_data_df_copy["duration_earth_days"]=""
for i, row in mars_temperature_data_df_copy.iterrows():
    mars_temperature_data_df_copy.loc[i,"duration_earth_days"]=(row["terrestrial_date"]-start_date).days
convert_to(mars_temperature_data_df_copy,7,7,"int64")
mars_temperature_data_df_copy.head(5)
```

```python
# Plot min temperature per days to identify peaks
plt.plot(mars_temperature_data_df_copy["duration_earth_days"],mars_temperature_data_df_copy["min_temp"])
plt.title("Min Temperature dependency from Terrestrial Days")
plt.ylabel("Minimum Temperature (Celsius)")
plt.xlabel("Number of Terrestrial Days")
plt.show()
```

```python
# Plot approximate year duration (having as an input hardcoded 537 and 1212 days count that we can visually identify from the above plot)
loc_df=mars_temperature_data_df_copy.loc[(mars_temperature_data_df_copy["duration_earth_days"]>500)&(mars_temperature_data_df_copy["duration_earth_days"]<1250),["duration_earth_days","min_temp"]]
x_start_point=[537 for i in range(len(loc_df["min_temp"]))]
x_end_point=[1212 for i in range(len(loc_df["min_temp"]))]
fig, ax = plt.subplots()
ax.scatter(loc_df["duration_earth_days"],loc_df["min_temp"],label="Min Temperature")
ax.plot(x_start_point,loc_df["min_temp"], color="r",label="approximate start point")
ax.plot(x_end_point,loc_df["min_temp"], color="purple",label="approximate end point")
ax.set_ylabel("Minimum Temperature (Celsius)")
ax.set_xlabel("Number of Terrestrial Days")
ax.set_title("Min Temperature dependency from Terrestrial Days (Martian year duration)")
ax.legend(loc="best")
plt.show()
```

Having analyzed data for 12 Martian months and 1867 Martian days we conclude the following:

On average, the third month has the coldest minimum temperature on Mars, and the eighth month is the warmest. But it is always very cold there in human terms!



Atmospheric pressure is, on average, lowest in the sixth month and highest in the ninth.

The distance from peak to peak is roughly 1212-537, or 675 days. A year on Mars appears to be about 675 days from the plot. Internet search confirms that a Mars year is equivalent to 687 earth days.
